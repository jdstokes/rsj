```python
    def rename_files(basedir,old_name,new_name):
        import os
        for fn in os.listdir(basedir):
          if not os.path.isdir(os.path.join(basedir, fn)):
              continue # Not a directory
          if old_name not in fn:
              continue # Already in the correct form
          new_name = fn.replace(old_name,new_name)
          os.rename(os.path.join(basedir, fn),
                os.path.join(basedir, new_name))
```

```python
for sub in subjects:
    
    if not os.path.isdir(os.path.join(mriDir,sub)):
        continue
    else:
        subjDir = os.path.join(mriDir,sub)
    for i,curDir in enumerate(oldDir_names):
        try:
            rename_files(subjDir,curDir,newDir_names[i])
        except ValueError:
            print "Oops!  That was no valid number.  Try again..."
```

##2) RS results <a id ='results_rs'></a>

**CA3 RS discrimination response**

    CA3 decreases pattern similarity when moving from ShapeM or ShapeN into ShapeT
    
    Lack of CA3 RS decrease predicts incorrect D-1 trials

**CA3 RS mainenance effect**

    When moving from ShapeT to ShapeN, high CA3 predicts correctness
    
    
**CA1 RS Maintenance**

    During processing of ShapeN information,i.e., moving from ShapeM to ShapeN and moving from ShapeN to ShapeN

**CA1 RS performance**

    High R CA1 predicts correct Same ShapeM and ShapeN trials but really high for Correct ShapeM and ShapeN D-1 trials trials
    
    Also high CA1 for correct ShapeN -> ShapeT
   
   

**CA3 RS performance**
    T<->M & N<->M: Really low CA3 for correct trials
    

    'Tpost: S,D1,D2'
    'Tpost correct: S,D1,D2'
    'Npost: S,D1,D2'
    'Npost correct: S,D1,D2'


    'D1 correct: Target, Morph, Novel'
    'D1pre: Target, Morph, Novel'
    'D1post: Target, Morph, Novel'
    'D1post correct: Target, Morph, Novel'

    'S CNorCM: CRT INCRT'
    'D1: CRT INCRT'
    'D1 CMpre: CRT INCRT'

    'D2 : CRT INCRT'
    'D2 CTpost: CRT INCRT'
    'D2 CNpost: CRT INCRT'
